```python
import numpy as np
expenses=np.array([1500,2000,1800,2200,1700,2500])
avg_expenses=expenses.mean() 
highest_expense=expenses.max()  
lowest_expense=expenses.min()  
total_expenses=expenses.sum()  
print("Average monthly expense:",avg_expenses)  #this returns the average of the expenses by dividing the total expenses by the number of months
print("Highest expense:",highest_expense)  #this returns the highest amount of expense
print("Lowest expense:",lowest_expense)   #this returns the lowest amount of expense
print("Total expenses:",total_expenses)  #this sums up all the expenses
print("First three months expenses:",expenses[0:3]) #this returns the expenses of the first three months
```

```python
import numpy as np
expenses=np.array([1500,2000,1800,2200,1700,2500])
avg_expenses=expenses.mean()
highest_expense=expenses.argmax()+1 #arg=argumentt(index)
lowest_expense=expenses.argmin()+1
total_expenses=expenses.sum()
print("Average monthly expense:",avg_expenses)  #this returns the average of the expenses by dividing the total expenses by the number of months
print("Highest expense month:",highest_expense) #this returns the month number with the highest expense
print("Lowest expense month:",lowest_expense) #this returns the month with the lowest expense
print("Total expenses:",total_expenses) #this returns the sum of all the expenses in the set
print("First three months expenses:",expenses[0:3]) #this returns the expenses of the first three months
```
