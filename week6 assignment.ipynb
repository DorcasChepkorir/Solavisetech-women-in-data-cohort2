Grouping data

```python
import pandas as pd

data={
  'Order_ID':   ['ORD001','ORD002','ORD003','ORD004','ORD005',
                'ORD006','ORD007','ORD008','ORD009','ORD010'], 
   'Product':    ['Laptop','Rice Bag','Headphones','Notebook','Blender',
                   'Rice Bag','Laptop','Notebook','Blender','Headphones'],
    'Category':   ['Electronics','Food','Electronics','Stationery','Appliances',
                  'Food','Electronics','Stationery','Appliances','Electronics'],
    'Region':     ['North','South','East','West','North',
                 'East','South','North','West','North'], 
    'Units_Sold': [5, 30, 12, 50, 8, 20, 3, 80, 6, 15],
    'Unit_Price': [850.00, 12.50, 45.00, 3.20, 65.00,
                   12.50, 850.00, 3.20, 65.00, 45.00],
    'Month':      ['January','January','February','February','March',
                   'March','April','April','May','May']
}

df=pd.DataFrame(data)
Units_Sold=df.groupby("Region")["Units_Sold"].sum()
print(Units_Sold)
```

```python
Unit_Price=df.groupby('Category')["Unit_Price"].mean()
print(Unit_Price)
```

```python
Category_summary=df.groupby('Category').agg({"Units_Sold":"sum","Unit_Price":"mean"})
print(Category_summary)
```

```python
Month_sales=df.groupby("Month")["Units_Sold"].sum()
print(Month_sales)
```

```python
Region_sum=df.groupby("Region")["Units_Sold"].sum()
Region_max=Region_sum.idxmax()
print(Region_max)
```

```python
Product_avg=df.groupby("Product")["Unit_Price"].mean()
Product_min=Product_avg.idxmin()
print(Product_min)
```

```python
Month_sum=df.groupby("Month")["Units_Sold"].sum()
Month_max=Month_sum.idxmax()
print(Month_max)
```

```python
Electronics_north=df[(df['Category']=="Electronics") & (df["Region"]=="North")]
Display=Electronics_north[["Product","Units_Sold","Unit_Price"]]
print(Display)   
```

```python
import pandas as pd

df=pd.DataFrame(data)
df["Total_Sales"]=df['Units_Sold']*df["Unit_Price"]
print(df)

Category_sum=df.groupby("Category")["Total_Sales"].sum()
Category_max=Category_sum.idxmax()
print("Category with highest total sales:",Category_max)
```
